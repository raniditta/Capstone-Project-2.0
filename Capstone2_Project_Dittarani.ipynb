# **2. Identifikasi Outlier**

Tujuan: Mengidentifikasi nilai ekstrem atau outlier dalam variabel numerik adalah langkah penting untuk memahami distribusi nilai dalam data. Outlier dapat mengindikasikan transaksi yang tidak biasa, kesalahan entri data, atau kasus khusus yang mungkin perlu diteliti lebih lanjut. Selain itu, jika outlier tidak ditangani, mereka dapat memengaruhi hasil analisis statistik atau model yang diterapkan, karena model statistik sering kali rentan terhadap pengaruh nilai yang sangat besar atau kecil. Mengetahui persentase outlier per produk memberikan wawasan apakah produk tertentu lebih sering mengalami anomali atau transaksi yang signifikan.

```python
# Mengidentifikasi outlier pada kolom numerik
num_cols = ['Discount', 'Profit', 'Quantity', 'Sales']
num_data = df[num_cols]

Q1 = num_data.quantile(0.25)
Q3 = num_data.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_positive_mask = (num_data > upper_bound).any(axis=1)
outliers_negative_mask = (num_data < lower_bound).any(axis=1)

df['Tipe_Outlier'] = 'Tidak Ada Outlier'
df.loc[outliers_positive_mask, 'Tipe_Outlier'] = 'Outlier Positif'
df.loc[outliers_negative_mask, 'Tipe_Outlier'] = 'Outlier Negatif'

outlier_persen_per_produk = df.groupby(['Product', 'Tipe_Outlier']).size().unstack(fill_value=0)
outlier_persen_per_produk = outlier_persen_per_produk.div(outlier_persen_per_produk.sum(axis=1), axis=0) * 100
print("Persentase dari setiap 'Tipe Outlier' untuk setiap 'Produk':")
print(outlier_persen_per_produk)
```

# **3. Pemeriksaan Distribusi Data**

Tujuan: Memahami apakah variabel kunci mengikuti distribusi normal sangat penting dalam memilih uji statistik yang tepat. Banyak metode analisis statistik (seperti ANOVA atau korelasi Pearson) mensyaratkan asumsi normalitas. Jika distribusi suatu variabel tidak normal, maka pendekatan lain, seperti transformasi data atau menggunakan uji non-parametrik, mungkin lebih sesuai. Pengecekan ini juga membantu dalam memahami karakteristik distribusi data penjualan, profit, diskon, dan jumlah unit, yang pada gilirannya dapat memengaruhi keputusan bisnis.

```python
# Menguji normalitas data pada beberapa kolom penting
columns_of_interest = ['Sales', 'Quantity', 'Discount', 'Profit']
alpha = 0.05

for column in columns_of_interest:
    data = df[column]
    stat, p_value = shapiro(data)
    print(f"\nUji Normalitas Shapiro-Wilk ({column}):")
    print(f"  p-value: {p_value}")
    if p_value > alpha:
        print(f"  Distribusi dari {column} kemungkinan normal.")
    else:
        print(f"  Distribusi dari {column} kemungkinan tidak normal.")

    plt.figure(figsize=(8, 6))
    sns.histplot(data, kde=True, color='skyblue', bins=30)
    plt.title(f'Distribusi dari {column}')
    plt.xlabel(column)
    plt.ylabel('Frekuensi')
    plt.show()
```

### Analisis Uji Normalitas Menggunakan Shapiro-Wilk

Hasil uji Shapiro-Wilk menunjukkan bahwa distribusi data untuk variabel **Sales**, **Quantity**, **Discount**, dan **Profit** kemungkinan besar tidak normal, terlihat dari p-value yang sangat rendah (di bawah 0.05). Berikut adalah ringkasan hasil uji normalitas:

#### 1. Distribusi Sales
   - **p-value**: `4.07 x 10^-105`
   - **Interpretasi**: Distribusi **Sales** tidak normal, mengindikasikan adanya outliers atau skewness, dengan beberapa nilai penjualan yang jauh lebih tinggi atau lebih rendah dari yang lain.

#### 2. Distribusi Quantity
   - **p-value**: `1.52 x 10^-65`
   - **Interpretasi**: Distribusi **Quantity** juga tidak normal, kemungkinan disebabkan oleh banyak transaksi dengan kuantitas rendah dan beberapa dengan kuantitas tinggi.

#### 3. Distribusi Discount
   - **p-value**: `1.03 x 10^-84`
   - **Interpretasi**: Distribusi **Discount** tidak normal, mungkin karena variasi dalam strategi diskon yang diterapkan, menghasilkan skewness.

#### 4. Distribusi Profit
   - **p-value**: `1.70 x 10^-107`
   - **Interpretasi**: Distribusi **Profit** tidak normal, disebabkan oleh transaksi dengan profit yang sangat tinggi atau rendah, berkaitan dengan variasi pada sales, quantity, dan discount.

# **5. Analisis Diskon dan Pengaruhnya terhadap Penjualan dan Profit**

Tujuan: Menganalisis pengaruh diskon terhadap penjualan dan margin profit adalah penting untuk menilai efektivitas strategi harga yang ada. Memberikan diskon yang terlalu besar bisa meningkatkan penjualan tetapi mengurangi profitabilitas, sedangkan diskon yang terlalu kecil mungkin tidak cukup menarik bagi pelanggan. Analisis ini membantu dalam menemukan keseimbangan antara diskon yang menarik bagi pelanggan dan yang tetap menguntungkan bagi perusahaan. Hasil ini bisa dijadikan panduan untuk menentukan batas diskon yang optimal agar perusahaan tetap mempertahankan profit margin yang sehat.

```python
# Menghitung margin profit dan menganalisis pengaruh diskon terhadap penjualan dan margin profit
df['Profit Margin'] = (df['Profit'] / df['Sales']) * 100
discount_performance = df.groupby('Discount').agg({'Sales': 'mean', 'Profit Margin': 'mean'}).reset_index()
discount_performance = discount_performance.sort_values(by='Discount', ascending=True)

# Plot pengaruh diskon pada penjualan dan margin profit
plt.figure(figsize=(12, 6))
sns.lineplot(data=discount_performance, x='Discount', y='Sales', marker='o', label='Penjualan Rata-Rata')
sns.lineplot(data=discount_performance, x='Discount', y='Profit Margin', marker='o', label='Margin Profit Rata-Rata')
plt.title('Pengaruh Diskon terhadap Penjualan dan Margin Profit')
plt.xlabel('Diskon (%)')
plt.ylabel('Nilai')
plt.legend()
plt.grid(True)
plt.show()
```

## Analisis Pengaruh Diskon terhadap Penjualan dan Margin Profit

Berdasarkan grafik yang menunjukkan pengaruh diskon terhadap penjualan dan margin profit, berikut adalah analisisnya:

## 1. Hasil Analisis
- **Penjualan Rata-Rata**:
  - Penjualan rata-rata meningkat seiring dengan peningkatan diskon, namun ada penurunan signifikan pada diskon sekitar 0,6.
  - Diskon **50%** menghasilkan penjualan tertinggi, menandakan daya tarik yang kuat pada level ini.

- **Margin Profit Rata-Rata**:
  - Margin profit rata-rata cenderung menurun dengan meningkatnya diskon. Diskon tinggi dapat meningkatkan penjualan, tetapi margin profit dapat menjadi negatif.
  - Diskon di kisaran **20%-30%** menunjukkan margin yang masih positif, menjadi batas optimal dalam memberikan diskon.

## 2. Kesimpulan
- **Optimalisasi Diskon**: Diskon hingga 20%-30% bisa meningkatkan penjualan tanpa mengurangi margin profit secara signifikan.
- **Diskon Tinggi vs Profitabilitas**: Diskon yang terlalu besar berisiko merugikan profitabilitas, sehingga harus dihindari kecuali untuk alasan khusus.


# **6. Analisis Nilai Lifetime Pelanggan (Customer Lifetime Value / CLV)**

Tujuan: Mengukur Customer Lifetime Value (CLV) adalah strategi yang berfokus pada nilai jangka panjang dari pelanggan. CLV membantu perusahaan mengidentifikasi pelanggan dengan nilai tertinggi, sehingga perusahaan dapat fokus pada upaya retensi untuk pelanggan bernilai tinggi. CLV juga memberikan panduan untuk alokasi anggaran pemasaran dan penjualan dengan lebih efisien. Misalnya, pelanggan yang memiliki nilai CLV tinggi mungkin mendapatkan prioritas dalam program loyalitas, diskon khusus, atau kampanye pemasaran yang lebih ditargetkan.

```python
# Asumsi periode retensi adalah 12 bulan untuk menghitung CLV
df['Periode Retensi'] = 12
df['CLV'] = df['Sales'] * df['Periode Retensi']

# Segmentasi CLV berdasarkan segmen pelanggan
clv_segment = df.groupby('Segment')['CLV'].mean().reset_index()
print("CLV Rata-Rata per Segmen:")
print(clv_segment)

# Visualisasi CLV Rata-Rata per Segmen
plt.figure(figsize=(8, 5))
sns.barplot(data=clv_segment, x='Segment', y='CLV', palette='viridis')
plt.title('CLV Rata-Rata per Segmen')
plt.ylabel('CLV Rata-Rata')
plt.xlabel('Segmen')
plt.xticks(rotation=45)
plt.show()
```

## 1. Tujuan Analisis

# **7. Analisis Sensitivitas Diskon terhadap Penjualan dan Profit**

Misalnya, jika diskon 10-20% meningkatkan penjualan tanpa menurunkan margin profit terlalu banyak, maka rentang tersebut mungkin merupakan pilihan diskon yang ideal. Dengan demikian, perusahaan dapat menetapkan strategi diskon yang berkelanjutan.

# **8. Clustering Berdasarkan Penjualan dan Profit**

Tujuan: Clustering digunakan untuk mengelompokkan produk atau transaksi berdasarkan karakteristik yang serupa, dalam hal ini penjualan dan profit. Tujuannya adalah untuk mengidentifikasi pola atau segmen produk yang mungkin memerlukan pendekatan pemasaran atau penjualan yang berbeda. Misalnya, produk dalam kelompok dengan profit tinggi dan penjualan rendah mungkin cocok untuk promosi khusus untuk meningkatkan volume, sementara produk dengan penjualan tinggi dan profit rendah bisa dievaluasi untuk efisiensi biaya. Segmentasi ini membantu dalam menyusun strategi produk yang lebih terarah dan optimal.

```python
# Normalisasi fitur dan menggunakan K-Means clustering
scaled_features = StandardScaler().fit_transform(df[['Sales', 'Profit']])
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)

# Visualisasi hasil clustering
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Cluster', palette='viridis')
plt.title('Hasil Clustering berdasarkan Penjualan dan Profit')
plt.xlabel('Penjualan')
plt.ylabel('Profit')
plt.legend(title='Cluster')
plt.show()
```

# **9. Analisis Variansi (ANOVA) untuk Menilai Perbedaan Profit Berdasarkan Segmen**

Tujuan: Uji ANOVA digunakan untuk mengevaluasi apakah terdapat perbedaan signifikan dalam profit rata-rata antara segmen pelanggan yang berbeda. Jika hasilnya signifikan, maka ada bukti bahwa segmen tertentu mungkin lebih menguntungkan daripada yang lain. Hal ini memberikan wawasan yang sangat berharga dalam menentukan prioritas segmen pelanggan untuk alokasi sumber daya dan pengembangan strategi pemasaran. Misalnya, jika segmen perusahaan lebih menguntungkan daripada individu, maka kampanye penjualan dapat lebih difokuskan pada perusahaan untuk mengoptimalkan profit.

```python
# Uji ANOVA untuk profit rata-rata per segmen
segments = [df['Profit'][df['Segment'] == seg] for seg in df['Segment'].unique()]
stat, p_value = f_oneway(*segments)

print("\nHasil Uji ANOVA Profit Berdasarkan Segmen:")
print(f"p-value: {p_value}")
if p_value < 0.05:
    print("Terdapat perbedaan signifikan dalam profit antar segmen pelanggan.")
else:
    print("Tidak terdapat perbedaan signifikan dalam profit antar segmen pelanggan.")
```

## Analisis Hasil Uji ANOVA Profit Berdasarkan Segmen

Berdasarkan hasil uji ANOVA yang dilakukan untuk melihat profit rata-rata per segmen, berikut adalah analisisnya:

## 1. Hasil Uji ANOVA
- **p-value** yang diperoleh dari analisis ini adalah **0.4074**. Dalam konteks statistik, nilai p-value ini akan dibandingkan dengan tingkat signifikansi yang umum digunakan, yaitu **0.05**, untuk menentukan apakah hasilnya signifikan atau tidak.
- Karena **p-value (0.4074) > 0.05**, kita tidak memiliki cukup bukti untuk mengatakan bahwa ada perbedaan signifikan dalam profit rata-rata di antara segmen-segmen pelanggan yang berbeda.

## 2. Kesimpulan
- Dari hasil analisis ini, dapat disimpulkan bahwa **tidak ada perbedaan signifikan dalam profit rata-rata antara segmen-segmen pelanggan**. Ini menunjukkan bahwa profit rata-rata yang dihasilkan dari segmen pelanggan **Enterprise**, **SMB**, dan **Strategic** cenderung cukup mirip.
- Dengan temuan ini, perusahaan bisa mempertimbangkan bahwa profit tidak terlalu dipengaruhi oleh segmentasi pelanggan. Oleh karena itu, jika perusahaan ingin meningkatkan profit, mungkin perlu untuk mempertimbangkan faktor lain di luar segmentasi, seperti strategi harga, jenis produk yang ditawarkan, atau efisiensi dalam operasional.
